````python
#2ì°¨ì› ìŠ¬ë¼ì´ì‹±
#row, col ê°ê°ì— ì¸ë±ì‹± ì ìš©

array1d=np.arange(start=1, stop=10)
array2d=array1d.reshape(3,3)
print('array2d:\n', array2d)
print('array2d[0:2, 0:2] \n', array2d[0:2, 0:2])
print('array2d[1:3, 0:3] \n', array2d[1:3, 0:3])
print('array2d[1:3, :] \n', array2d[1:3, :])
print('array2d[:, :] \n', array2d[:, :])
print('array2d[:2, 1:] \n', array2d[:2, 1:])
print('array2d[:2, 0] \n', array2d[:2, 0])

print(array2d[0])
print(array2d[1])
print('array2d[0] shape:', array2d[0].shape, 'array2d[1] shape:', array2d[1].shape)
````

**íŒ¬ì‹œ ì¸ë±ì‹±(Fancy Indexing)** \
ë¦¬ìŠ¤íŠ¸ë‚˜ ndarrayë¡œ ì¸ë±ìŠ¤ ì§‘í•©ì„ ì§€ì •í•˜ë©´ í•´ë‹¹ ìœ„ì¹˜ì˜ ì¸ë±ìŠ¤ì— í•´ë‹¹í•˜ëŠ” ndarrayë¥¼ ë°˜í™˜í•˜ëŠ” ì¸ë±ì‹± ë°©ì‹

````python
#Fancy Indexing

#2ì°¨ì› ndarrayì— ì ìš©
array1d=np.arange(start=1, stop=10)
array2d=array1d.reshape(3,3)
array3=array2d[[0,1],2]
print('array2d[[0,1],2]=>', array3.tolist())
array4=array2d[[0,1],0:2]
print('array2d[[0,1],0:2]=>', array4.tolist())
array5=array2d[[0,1]]
print('array2d[[0,1]]=>', array5.tolist())
````

**ë¶ˆë¦° ì¸ë±ì‹±(Boolean Indexing)** \
ì¡°ê±´ í•„í„°ë§ê³¼ ê²€ìƒ‰ì„ ë™ì‹œì— í•  ìˆ˜ ìˆì–´ ìì£¼ ì‚¬ìš©ë˜ëŠ” ì¸ë±ì‹± ë°©ì‹\
[ ] ë‚´ì— ì¡°ê±´ë¬¸ ê¸°ì¬

````python
#Boolean Indexing
array1d=np.arange(start=1, stop=10)
#[]ì•ˆì— array1d>5 Boolean indexingì„ ì ìš©
array3=array1d[array1d>5]
print('array1d>5 ë¶ˆë¦° ì¸ë±ì‹± ê²°ê³¼ ê°’ : ', array3)

boolean_indexes=np.array([False, False, False, False, False,  True,  True,  True,  True])
array3=array1d[boolean_indexes]
print('ë¶ˆë¦° ì¸ë±ìŠ¤ë¡œ í•„í„°ë§ ê²°ê³¼:', array3)

indexes=np.array([5,6,7,8])
array4=array1d[indexes]
print('ì¼ë°˜ ì¸ë±ìŠ¤ë¡œ í•„í„°ë§ ê²°ê³¼:', array4)
````

**ì •ë ¬ëœ í–‰ë ¬ì˜ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜í•˜ê¸° - np.argsort()**
* ì›ë³¸ í–‰ë ¬ì´ ì •ë ¬ë˜ì—ˆì„ ë•Œ ê¸°ì¡´ ì›ë³¸ í–‰ë ¬ì˜ ì›ì†Œì— ëŒ€í•œ ì¸ë±ìŠ¤ë¥¼ í•„ìš”ë¡œ í•  ë•Œ ì´ìš©
* ì •ë ¬ í–‰ë ¬ì˜ ì›ë³¸ í–‰ë ¬ ì¸ë±ìŠ¤ë¥¼ ndarray í˜•ìœ¼ë¡œ ë°˜í™˜
* ***ë„˜íŒŒì´ì—ì„œ ë§¤ìš° í™œìš©ë„ê°€ ë†’ìŒ***
  * ndarrayëŠ” tableì´ë‚˜ Dataframeê³¼ ê°™ì€ ë©”íƒ€ ë°ì´í„°ë¥¼ ê°€ì§ˆ ìˆ˜ ì—†ìŒ

````python
#ì •ë ¬ëœ í–‰ë ¬ì˜ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜í•˜ê¸°

org_array=np.array([3,1,9,5])
sort_indices=np.argsort(org_array)
print(type(sort_indices))
print('í–‰ë ¬ ì •ë ¬ ì‹œ ì›ë³¸ í–‰ë ¬ì˜ ì¸ë±ìŠ¤:', sort_indices)

org_array=np.array([3,1,9,5])
sort_indices_desc=np.argsort(org_array)[::-1]
print('í–‰ë ¬ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ ì‹œ ì›ë³¸ í–‰ë ¬ì˜ ì¸ë±ìŠ¤:', sort_indices_desc)

#í•™ìƒë³„ ì‹œí—˜ì„±ì ì„ ë°ì´í„°ë¡œ í‘œí˜„-> ì‹œí—˜ ì„±ì ìˆœìœ¼ë¡œ í•™ìƒ ì´ë¦„ ì¶œë ¥
name_array=np.array(['John','Mike', 'Sarah', 'Kate', 'Samuel'])
score_array=np.array([78,95,84,98,88])
sort_indices_asc=np.argsort(score_array)
print('ì„±ì  ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬ ì‹œ score_arrayì˜ ì¸ë±ìŠ¤:', sort_indices_asc)
print('ì„±ì  ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ name_arrayì˜ ì´ë¦„ ì¶œë ¥:', name_array[sort_indices_asc])
````
