(Define 'function' to solve this probelm)

```python
users['gender'] = users['gender'].apply(lambda x : 1 if x=='M' else 0 )
a=users.groupby('occupation')['gender'].sum()/users['occupation'].value_counts()
a.sort_values(ascending=False)
```
